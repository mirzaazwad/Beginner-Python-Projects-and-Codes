#### Import Libraries

```python
from motionchart.motionchart import MotionChart
import pandas as pd
import numpy as np
```

#### Making the dataframe using concat

```python
df1=pd.read_excel('energy_data.xlsx','VIC')
df2=df=pd.read_excel('energy_data.xlsx','TAS')
df3=df=pd.read_excel('energy_data.xlsx','QLD')
df4=df=pd.read_excel('energy_data.xlsx','SA')
df5=df=pd.read_excel('energy_data.xlsx','WA')
df6=df=pd.read_excel('energy_data.xlsx','NT')
df7=df=pd.read_excel('energy_data.xlsx','NSW')
df=pd.concat([df1,df2,df3,df4,df5,df6,df7],ignore_index=True)
df.replace(np.NaN,0,inplace=True)
df
```

### Now let's look at the relationship between all variables impacting the energy generation over time. Ensure that you have combined all the data from the different states. Ensure that your data is aggregated by year, state, the total energy produced (total_production), and has a separate column for each of the fuel types.


#### Modifying the dataframe to produce the motionchart

```python
years=[i for i in range(2009,2019)]
df2=pd.melt(df,id_vars=['State','Fuel_Type','Category'],var_name='Year',value_name='Fuel_Generated',value_vars=years)
data=[]
state=['VIC','TAS','QLD','SA','WA','NT','NSW']
year=np.arange(2009,2019,1)
total_production=[]
for i in year:
    for j in state:
        total_production.append(df[df['State']==j][i].sum())
total_production=np.array(total_production, dtype='float64')
for year in range(2009,2019):
    data_tmp=pd.DataFrame({'State':['VIC','TAS','QLD','SA','WA','NT','NSW'],
                 'Year':[year for i in range(7)],
                  'Black_Coal': np.array(df2[(df2.Fuel_Type=='Black coal') & (df2.Year==year)]['Fuel_Generated']),
                  'Brown_Coal':np.array(df2[(df2.Fuel_Type=='Brown coal') & (df2.Year==year)]['Fuel_Generated']),
                  'Natural_Gas':np.array(df2[(df2.Fuel_Type=='Natural gas') & (df2.Year==year)]['Fuel_Generated']),
                   'Oil_Products':np.array(df2[(df2.Fuel_Type=='Oil products') & (df2.Year==year)]['Fuel_Generated']),
                   'Other':np.array(df2[(df2.Fuel_Type=='Other') & (df2.Year==year)]['Fuel_Generated']),
                  'Biomass':np.array(df2[(df2.Fuel_Type=='Biomass') & (df2.Year==year)]['Fuel_Generated']),
                  'Wind':np.array(df2[(df2.Fuel_Type=='Wind') & (df2.Year==year)]['Fuel_Generated']),
                  'Hydro':np.array(df2[(df2.Fuel_Type=='Hydro') & (df2.Year==year)]['Fuel_Generated']),
                  'Large-Scale_Solar_PV':np.array(df2[(df2.Fuel_Type=='Large-scale solar PV') & (df2.Year==year)]['Fuel_Generated']),
                  'Small-Scale_Solar_PV':np.array(df2[(df2.Fuel_Type=='Small-scale solar PV') & (df2.Year==year)]['Fuel_Generated']),
                  'Geothermal':np.array(df2[(df2.Fuel_Type=='Geothermal') & (df2.Year==year)]['Fuel_Generated'])
                  })
    data.append(data_tmp)
data=pd.concat(data,ignore_index=True)
data['total_production']=total_production
data
```
