#### Loading the CSV file and making the dataframe

```python
df=pd.read_csv('twitter_data.csv')
df
```

### We can observe a relationship between the number of followers a person has and the number of lists the author has been listed in

```python
plt.scatter(df['nFollowers'],df['nListed'])
slope,intercept,r_value,p_value,std_err=linregress(df['nFollowers'],df['nListed'])
line=[slope*x+intercept for x in df['nFollowers']]
plt.plot(df['nFollowers'],line,color='r')
plt.xlabel('Followers a person has')
plt.ylabel('The Number of Lists the person is in')
plt.suptitle('Corelation Between Number of lists a person is in \n and the Number of Followers the person has')
plt.savefig('Task B2')
```

```python
r_value #The r_value or the corelation point being closer to +
#1, indicates that there is significant positive corelation
```

```python
p_value #p_value less than 0.05 indicates that there is significant corelation
```
